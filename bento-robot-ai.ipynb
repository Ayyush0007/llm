```python
from pathlib import Path

dataset_path = Path('/kaggle/working/idd-detection1-1')

for split in ['train', 'valid']:
    imgs = list((dataset_path / split / 'images').glob('*.*'))
    labels = list((dataset_path / split / 'labels').glob('*.txt'))
    print(f'{split}: {len(imgs)} images, {len(labels)} labels')

print("\nDataset ready for training!")
```

```python
from ultralytics import YOLO
import glob, shutil

# Load base model — pretrained on COCO (80 classes)
model = YOLO('yolov8n.pt')

# Fine-tune on Indian roads data
model.train(
    data='/kaggle/input/self-driving-car/data.yaml',
    epochs=50,
    imgsz=640,
    batch=32,
    device='0,1',
    name='bento-v1',
    patience=15,
    augment=True,
    mosaic=1.0,
    lr0=0.01,
    save_period=10
)

print("Training complete!")

# Find and save the weights
weights = glob.glob('/kaggle/working/runs/**/best.pt', recursive=True)
if weights:
    shutil.copy(weights[0], '/kaggle/working/bento_model.pt')
    print(f"✅ Model saved to /kaggle/working/bento_model.pt")
else:
    print("❌ Weights not found")
```

```python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt

# Load pretrained YOLOv8 model (downloads automatically)
model = YOLO('yolov8n.pt')

# Test on a sample image from the repo
import os

# Find any image in the repo
img_path = None
for root, dirs, files in os.walk('/kaggle/working/Self-Driving-Car-on-Indian-Roads'):
    for f in files:
        if f.endswith(('.jpg', '.jpeg', '.png')):
            img_path = os.path.join(root, f)
            break
    if img_path:
        break

print("Found image:", img_path)

# Run detection
results = model(img_path)
results[0].show()

# Print what it detected
for box in results[0].boxes:
    cls = results[0].names[int(box.cls)]
    conf = float(box.conf)
    print(f"Detected: {cls} — confidence: {conf:.2f}")
```

```python
# Generate a test image using Python directly — no download needed
import numpy as np
from PIL import Image
import requests

# Use ultralytics built-in test assets (bundled locally)
from ultralytics.utils import ASSETS
print("Available test images:", list(ASSETS.glob("*")))

# Run on built-in bus.jpg that comes with ultralytics
results = model(ASSETS / "bus.jpg")

for box in results[0].boxes:
    cls = results[0].names[int(box.cls)]
    conf = float(box.conf)
    print(f"Detected: {cls} — confidence: {conf:.2f}")

results[0].save(filename="result.jpg")
print("Done! Check result.jpg in output")
```
